```python
meta = meta[meta['user'].isin(subdirs)]
```

```python
medians1 = (meta.iloc[:,2:7] >= meta.iloc[:,2:7].median()).astype('int')
medians2 = (meta.iloc[:,2:7] > meta.iloc[:,2:7].median()).astype('int')
```

```python
median1_scores = pd.merge(meta['user'], medians1, left_index=True, right_index=True)
median2_scores = pd.merge(meta['user'], medians2, left_index=True, right_index=True)
```

```python
median1_scores
```

```python
median2_scores
```

```python
labels = median1_scores.columns.values[1:]
labels
```

```python
# create new df with user column
merged = pd.DataFrame()
merged['user'] = median2_scores['user'].copy()

# greater or equal = 1, only greater = 0

for l in labels:
    one1 = (median1_scores[l] == 1).sum()
    zero1 = (median1_scores[l] == 0).sum()
    minc1 = min(one1, zero1)
    majc1 = max(one1, zero1)
    one2 = (median2_scores[l] == 1).sum()
    zero2 = (median2_scores[l] == 0).sum()
    minc2 = min(one2, zero2)
    majc2 = max(one2, zero2)
    r1 = minc1 / majc1
    r2 = minc2 / majc2
    if r1 > r2:
        print(f"{l}, r1: {round(r1, 3)}, r2: {round(r2, 3)}, so r1: greater or equal")
        
        merged[l] = median1_scores[l].copy()

    else:
        print(f"{l}, r1: {round(r1, 3)}, r2: {round(r2, 3)}, so r2: only greater than")

        merged[l] = median2_scores[l].copy()
```
