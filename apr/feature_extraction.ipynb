```python
from joblib import load, dump

file_info = load('/media/emile/disk/chatbot_10_15_lengths_paths.pkl')
files = [x[2] for x in file_info]
files
```

```python
import opensmile

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
raw_features = smile.process_files(files)

dump(raw_features, '/media/emile/disk/chatbot_10_15_raw_features.pkl')
```

```python
raw_features
```

```python
import pandas as pd
```

```python
features = raw_features.copy()
```

```python
file_names = [x[0] for x in file_info]
```

```python
df_names = pd.DataFrame(file_names, columns=['ID'])
```

```python
features.reset_index(drop=True, inplace=True)
```

```python
features
```

```python
df = pd.merge(df_names, features, left_index=True, right_index=True)
```

```python
df
```

```python
df['Group'] = (df['ID'].str.split('_').str[:3]).str.join('_')
```

```python
df['ID'] = df['ID'].str[:-4]
```

```python
col_name = 'Group'
col = df.pop(col_name)
df.insert(1, col_name, col)
```

```python
lengths = [x[1] for x in file_info]
```

```python
df['Length'] = pd.DataFrame(lengths, columns=['Length'])
```

```python
len(set(df['Group']))
```

```python
df.to_pickle('../analysis/temp_chatbot_10_15.pkl')
```

```python
scores = pd.read_csv('../analysis/personality_scores.csv')
```
